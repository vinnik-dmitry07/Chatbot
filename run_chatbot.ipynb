```
from pathlib import Path

GDRIVE_ROOT = Path('/content/drive/MyDrive/')
SAVE_DIR = GDRIVE_ROOT / 'chatbot_model'
DATA_DIR = GDRIVE_ROOT / 'chatbot_data'
```

```
from google.colab import drive

drive.mount(str(GDRIVE_ROOT.parent))
```

```
from parlai.scripts.display_model import DisplayModel

DisplayModel.main(
    model_file=str(SAVE_DIR / 'model'),  # or 'model.checkpoint'
    task='jsonfile',
    jsonfile_datapath=DATA_DIR / 'data',
    jsonfile_datatype_extension=True,
    skip_generation=False,
    num_examples=1000,
)
```

```
from parlai.scripts.interactive import Interactive

Interactive.main(model_file=str(SAVE_DIR / 'model'))
```
